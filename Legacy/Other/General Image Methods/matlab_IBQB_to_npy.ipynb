```python
import numpy as np

# Example list of arrays with different shapes
frames = [np.random.rand(100, 100), np.random.rand(200, 150), np.random.rand(120, 120)]

# Convert the list to a NumPy array of objects
frames_as_objects = np.array(frames, dtype=object)

# Save the array of objects
np.save('frames.npy', frames_as_objects)

# Load the data back
loaded_frames = np.load('frames.npy', allow_pickle=True)

# Verify the shapes
for i, frame in enumerate(loaded_frames):
    print(f"Frame {i} shape: {frame.shape}")
```
