```python
import cv2

import numpy as np

video_path='/Users/henryschnieders/Desktop/Research/My_work/Data/inorganic.MOV'

pixel_dimensions=90
desired_frames=500
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

 

cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

# Initialize an empty list to store frames
frames = []

# Frame processing loop
frame_count = 0
while cap.isOpened() and frame_count < desired_frames:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Resize frame to the desired dimensions
    frame_resized = cv2.resize(frame, (pixel_dimensions, pixel_dimensions))

    
    # Add the frame to the list
    frames.append(frame_resized)
    

    frame_count += 1
cap.release()


lsit=[]
for i in range(len(frames)):
    frame = frames[i]
    frame = cv2.flip(frame, 1) #normal camera feed is mirrored
    frame = cv2.resize(frame, (90, 90))
    frame=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    print(f'working on frame {i}')

    lsit+=[frame]
    
np.save('/Users/henryschnieders/Desktop/Research/My_work/Real_time_recognition/All_face_sections/amplitude_heatmap/Results/resized_inorganic.npy', lsit)
```
