```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the sine wave
sampling_rate = 1000  # Sampling rate in Hz
T = 1.0 / sampling_rate  # Sampling interval
t = np.arange(0.0, 1.0, T)  # Time vector (1 second)

freq = 50  # Frequency of the sine wave in Hz
amplitude = 1.0  # Amplitude of the sine wave

# Generate the sine wave
sine_wave = amplitude * np.sin(2 * np.pi * freq * t)

# Compute the FFT
fft_result = np.fft.fft(sine_wave)

# Compute the frequencies corresponding to the FFT result
fft_freqs = np.fft.fftfreq(len(sine_wave), T)

# Plot the sine wave
plt.figure()
plt.plot(t, sine_wave)
plt.title('Sine Wave')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# Plot the magnitude of the FFT
plt.figure()
plt.plot(fft_freqs, np.abs(fft_result))
plt.title('FFT of Sine Wave')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.xlim(0, sampling_rate / 2)  # Only plot the positive frequencies
plt.grid()

plt.show()
```
